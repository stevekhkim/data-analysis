```python
# 1월 데이터 전처리
def preprocess_01(df):
  
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-01'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df


# 2월 데이터 전처리
def preprocess_02(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-02'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df


# 3월 데이터 전처리
def preprocess_03(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-03'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df


# 4월 데이터 전처리
def preprocess_04(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-04'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df


# 5월 데이터 전처리
def preprocess_05(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-05'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df


# 6월 데이터 전처리
def preprocess_06(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-06'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df


# 7월 데이터 전처리
def preprocess_07(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-07'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df


# 8월 데이터 전처리
def preprocess_08(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-08'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df
  

# 9월 데이터 전처리
def preprocess_09(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-09'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df


# 10월 데이터 전처리
def preprocess_10(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-10'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df


# 11월 데이터 전처리
def preprocess_11(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-11'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df


# 12월 데이터 전처리
def preprocess_12(df):
    
  df = df.drop('측정망', axis=1)
  df = df.melt(id_vars='측정소명')

  df.columns = ['region', 'day', 'pm10']

  df['region'] = df['region'].str.replace('[',"")
  df['region'] = df['region'].str.split(']', expand=True)[0]

  df['month'] = '2019-12'  ## 월정보만 임의로 변경해주면 됩니다.
  df['day'] = df['day'].str.replace('일','')
  df['date'] = df['month'] + '-' + df['day']
  df['date'] = pd.to_datetime(df['date'])
  df = df.drop(['day','month'], axis=1)

  df = df.groupby(['region', 'date'])['pm10'].mean().reset_index()
  df = df.rename(columns={'date':'day'})

  return df
```
